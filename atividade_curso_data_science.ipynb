```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('populacao_brasileira.csv', sep=';')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
df.plot(x='ano', y='population', kind='bar', ax=axes[0], title='Crecimento da populacao brasileira(Barras)', color='g')
df.plot(x='ano', y='population', kind='line', ax=axes[1], title='Crecimento da populacao brasileira(linhas)', color='b')
plt.tight_layout()
plt.show()
```

```python
df.plot(x='ano', y='population', kind='bar', title='Crescimento da populao brasileira', color='g')
df.plot(kind='line', ax=plt.gca(), color='b')
plt.show()
```
